```python
import csv
import pandas as df.sum(axis = 0, skipna = True) pd
```

```python
daily = pd.read_csv('/Users/fernieqin/Desktop/DSC106/hw2/daily_sales.csv').drop(columns='Day')
```

```python
daily = daily.rename(columns={'Unnamed: 0':'Day'})
```

```python
daily_day = daily.groupby('Day').mean()
daily_day.head()
```

```python
daily_day['HM-total'] = sum(daily_day[i] for i in daily_day.columns if 'HM-' in i)
daily_day['CF-total'] = sum(daily_day[i] for i in daily_day.columns if 'CF-' in i)
daily_day['FF-total'] = sum(daily_day[i] for i in daily_day.columns if 'FF-' in i)
daily_day.head()
```

```python
daily_day['NE-total'] = sum(daily_day[i] for i in daily_day.columns if '-NE' in i)
daily_day['SW-total'] = sum(daily_day[i] for i in daily_day.columns if '-SW' in i)
daily_day['NW-total'] = sum(daily_day[i] for i in daily_day.columns if '-NW' in i)
daily_day['SE-total'] = sum(daily_day[i] for i in daily_day.columns if '-SE' in i)
daily_day['C-total'] = sum(daily_day[i] for i in daily_day.columns if '-C' in i)
daily_day.head()
```

```python
#sort the days
new_index = pd.Series([5, 1, 6, 7,4,2,3])
daily_day = daily_day.set_index([new_index])
daily_day.sort_index(axis = 0) 
```
